```python

############################################################################ Power density

p = 3*10**4
tag_default['laser'] = 5.76*(p/100)
# tag_default['laser'] = p*1.23 # (W/cm^2) to s^-1



# Er: conc, Yb=0.2, nanoparticle diameter: 8 nm, critical distance: 1 nm

# Er_conc=[0.02, 0.04, 0.06, 0.08, 0.1, 0.12, 0.15, 0.2, 0.25, 0.3, 0.35, 0.4, 0.45, 0.5, 0.55, 0.6, 0.65, 0.7, 0.75, 0.8]

Er_conc=[0.2]

Yb_conc=0.2
all_results = []
ratios=[]
d=4
r0=1
t1 = 2000
t2 = 5000


for c in Er_conc:
    
    my_lattice = Lattice(Yb_conc, c, d, r0)
    my_simulator = Simulator(my_lattice, tag = tag_default)
    r = my_simulator.simulate(t1, t2)


    # multiply 10^6 from us to s
    Red = r['red_avg']*10**6
    Green = r['green_avg']*10**6
    Green50s = r['green50_avg']*10**6
    Green60s = r['green60_avg']*10**6
    
    results = {}
    results[c] = Red, Green, Green50s, Green60s
    all_results.append(results)

    ratio={}
    ratio[c] = Red / Green
    ratios.append(ratio)



print(all_results)
print(ratios)

for key, value in r.items():
    print(f'Key: {key}, Value: {value}')
```
