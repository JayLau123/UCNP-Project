```python
import numpy as np
import plotly.graph_objs as go
import pickle
```

```python
w_list = [100, 1000, 10000, 100000, 1000000]
```

```python
data_trials = []

for i in range(6, 12):
    filepath = f'myC_09_19_2024_{i}.pkl'
    with open(filepath, 'rb') as f:
        data_6 = pickle.load(f)
        data_trials.append(data_6)
```

```python

time = np.linspace(0, 6000, 60)
traces = []
color_map = {
    'bright purple': 'rgb(178, 102, 255)',
    'bright blue': 'rgb(0, 0, 255)',
    'bright green': 'rgb(0, 255, 0)',
    'orange': 'rgb(255, 153, 51)',
    'red': 'rgb(255, 0, 0)'
}
color_list = list(color_map.values())

for i, w in enumerate(w_list):
    color = color_list[i % len(color_list)]
    
    for j, data_6 in enumerate(data_trials):
        combined_data = (np.array(data_6[0.1][w]['er_distribution'][5]) + 
                         np.array(data_6[0.1][w]['er_distribution'][6]))
        
        downsampled = np.mean(np.reshape(combined_data, (60, -1)), axis=1)
        m = np.max(downsampled)
        percentage = (downsampled / m)
        
        traces.append(go.Scatter(x=time, y=percentage, mode='markers',
                                 marker=dict(color=color, size=3),  # Adjust the size parameter
                                 showlegend=(j == 0),  # Show legend only for the first trace of each `w`
                                 name=f'w={w}'))

fig = go.Figure(traces)
fig.update_layout(xaxis_title='Time', yaxis_title='Normalized population (a.u.)',
                  yaxis=dict(type='log', tickvals=[0.001, 0.01, 0.1, 1]),
                  width=800, height=600)  # Set width and height to make the plot square
fig.show()
```
